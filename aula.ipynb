```python
from selenium import webdriver
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://google.com.br")
```

```python
driver.title
```

```python
driver.current_url
```

```python
driver.fullscreen_window()
```

```python
driver.get_window_rect()
#informações atuais de como está a dimensão da página e a sua posição no eixo X e Y
```

```python
driver.set_window_rect(x=1000, y=100, height=800, width=800)
#podemos setar e dimensionar o tamanho da tela
```

```python
driver.maximize_window()
#só maximiza mas não deixa full screen
```

```python
driver.name
```

```python
driver.page_source
#podemos buscar todo o html da pagina
```
